```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/used_cars_messy.csv')
df=pd.DataFrame(df)
df.head()
```

Task 1

```python
print(df.info())
print(df.describe())
print(df.shape)
```

Task :2

```python
#drop rows where selling price is missing
df=df.dropna(subset=['selling_price'])
df
```

```python
#remove duplicate rows
df=df.drop_duplicates()
df
```

```python
#CLEAN BRAND COLUMN
df['brand']=df['brand'].str.strip().str.lower()
df
```

```python
#clean mileage
df['mileage']=df['mileage'].str.extract('(\d+\.?\d*)')
df['mileage']=df['mileage'].astype(float)
```

```python
#impute missing valies in input feature
df['mileage'].fillna(df['mileage'].mean(), inplace=True)
df
```

task 3

```python
mean_price = df['selling_price'].mean()
```

```python
y_pred = [mean_price] * len(df)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(df['selling_price'], y_pred)
```

```python
print("Baseline MAE:", mae)
```
