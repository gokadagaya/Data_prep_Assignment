Task-1
Clean the Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)
```

```python
df = df.drop_duplicates()
print(df)
```

Task-2
Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

df = pd.get_dummies(df, columns=['city'])

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print(df)
```

Task-3
Scale Numerical Features


```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df_minmax = df.copy()
df_standard = df.copy()

cols = ['age', 'annual_income']

mm = MinMaxScaler()
df_minmax[cols] = mm.fit_transform(df_minmax[cols])


ss = StandardScaler()
df_standard[cols] = ss.fit_transform(df_standard[cols])

# Print results
print("Min-Max Scaling:\n", df_minmax)
print("\nStandardization:\n", df_standard)
```

Min-Max Scaling converts values between 0 and 1, so it’s useful when you want all features in the same fixed range (like for neural networks or distance-based models).
Standardization converts data to have mean = 0 and standard deviation = 1, and it works better when data has outliers or is normally distributed.
